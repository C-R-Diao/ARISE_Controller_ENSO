```python
#============================================#
# functions for plots
#============================================#

#===== func: plot NAO =====#
def plot_nao(nino, hlt_range=[-0.5,0.5], title = None, plot_name = None):
    """
    plot annual mean NAO
    hlt_range: [lower_bound, higher_bound]; decide which region to fill
    """
    fig = plt.figure(figsize=(20, 20))
    ax = fig.subplots(5, 2)
    fig.suptitle("standardized NAO from SAI", size=40)
    
    for i in range(10):
        tmp = ax[i//2,i%2]
        tmp.plot(nino[i,:])
        total_yr = len(nino[i,:])
        
        #reference line and highlight
        tmp.axhline(y = 0, c = 'k', linestyle='-')
        tmp.axhline(y = hlt_range[0], c = 'k', linestyle='--')
        tmp.axhline(y = hlt_range[1], c = 'k', linestyle='--')
        dumy_x = nino[i,:].copy()
        dumy_x[dumy_x> hlt_range[0]] = 'nan'
        tmp.fill_between(np.arange(total_yr), dumy_x, hlt_range[0], color="b")
        dumy_x = nino[i,:].copy()
        dumy_x[dumy_x< hlt_range[1]] = 'nan'
        tmp.fill_between(np.arange(total_yr), dumy_x, hlt_range[1], color="r")

        # polish
        tmp.set_xlim(0,total_yr)
        #tmp.set_ylim(-2.5, 2.5)
        tmp.set_ylabel('NAO', size=30)
        
        tmp.set_xticks(np.arange(0, total_yr+1, 5))
        tmp.set_xticklabels(np.arange(2035,2071, 5))
        tmp.set_xlabel('month', size=30)

        tmp.tick_params(axis= 'both',labelsize= 20)
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    if plot_name != None:
        fig.savefig(plot_name + '.pdf', format='pdf', dpi=300)


#===== finc: plot_iv =====#
def plot_iv(nino, hlt_range=[-0.5,0.5], title = None, save_fig = None):
    """
    plot monthly mean IV indices
    hlt_range: [lower_bound, higher_bound]; decide which region to fill
    """
    fig = plt.figure(figsize=(20, 20))
    ax = fig.subplots(5, 2)
    if title != None:
        fig.suptitle(title, size=40)
    
    for i in range(10):
        tmp = ax[i//2,i%2]
        tmp.plot(nino[i,:])
        total_month = len(nino[i,:])
        
        #reference line and highlight
        tmp.axhline(y = 0, c = 'k', linestyle='-')
        tmp.axhline(y = hlt_range[0], c = 'k', linestyle='--')
        tmp.axhline(y = hlt_range[1], c = 'k', linestyle='--')
        dumy_x = nino[i,:].copy()
        dumy_x[dumy_x> hlt_range[0]] = 'nan'
        tmp.fill_between(np.arange(total_month), dumy_x, hlt_range[0], color="b")
        dumy_x = nino[i,:].copy()
        dumy_x[dumy_x< hlt_range[1]] = 'nan'
        tmp.fill_between(np.arange(total_month), dumy_x, hlt_range[1], color="r")

        # polish
        tmp.set_xlim(0,total_month)
        #tmp.set_ylim(-2.5, 2.5)
        tmp.set_ylabel('ONI', size=30)
        
        tmp.set_xticks(np.arange(0, total_month+1, 60))
        tmp.set_xticklabels(np.arange(2035,2071, 5))
        tmp.set_xlabel('momth', size=30)

        tmp.tick_params(axis= 'both',labelsize= 20)
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    if save_fig != None:
        fig.savefig(save_fig + '.pdf', format='pdf', dpi=300)
        
        
#===== func: plot gmst =====#
def plot_gmst(nino, plot_name = None):
    """
    hlt_range: [lower_bound, higher_bound]; decide which region to fill
    
    """
    fig = plt.figure(figsize=(20, 20))
    ax = fig.subplots(5, 2)
    fig.suptitle("GMST (ann) from SAI", size=40)
    
    for i in range(10):
        tmp = ax[i//2,i%2]
        tmp.plot(nino[i,:])
        total_yr = len(nino[i,:])
        
        #reference line
        tmp.axhline(y = 0, c = 'k', linestyle='-')

        # polish
        tmp.set_xlim(0,total_yr)
        #tmp.set_ylim(-2.5, 2.5)
        tmp.set_xticks(np.arange(0, total_yr+1, 5))
        tmp.set_xticklabels(np.arange(2035,2075, 5))
        tmp.set_xlabel('year', size=30)
        tmp.set_ylabel('GMST', size=30)
        tmp.tick_params(axis= 'both',labelsize= 20)
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    if plot_name != None:
        fig.savefig(plot_name + '.pdf', format='pdf', dpi=300)
        
        
#===== func: plot sai =====#
def plot_sai(s, plot_name = None):

    fig = plt.figure(figsize=(20, 20))
    ax = fig.subplots(5, 2)
    fig.suptitle("Sulfate (Tg) from SAI", size=40)
    
    for i in range(10):
        tmp = ax[i//2,i%2]
        tmp.plot(s[i,:])
        total_yr = len(s[i,:])

        # polish
        
        #reference line
        #tmp.axhline(y = 0, c = 'k', linestyle='-')
        
        tmp.set_xlim(0,total_yr)
        #tmp.set_ylim(0, 12)
        tmp.set_xticks(np.arange(0, total_yr+1, 5))
        #tmp.set_xticklabels(np.arange(2035,2075, 5))
        tmp.set_xlabel('year', size=30)
        tmp.set_ylabel("Sulfate (Tg)", size=20)
        tmp.tick_params(axis= 'both',labelsize= 20)
    
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    if plot_name != None:
        plot_path = '/glade/u/home/dchenrui/DARPA/figs/'
        fig.savefig(plot_path + plot_name + '.pdf', format='pdf', dpi=300)
```

```python
#============================================#
# Read data
#============================================#

#===== from ARISE =====#
PATH = "/glade/campaign/cesm/collections/ARISE-SAI-1.5/"
CASE = "b.e21.BW.f09_g17.SSP245-TSMLT-GAUSS-DEFAULT."
 
time_span = ['2035', '2069']
sst = read_data("SST", PATH, CASE, time_span)
ts = read_data("TREFHT", PATH, CASE, time_span)
psl = read_data("PSL", PATH, CASE, time_span)


#===== from SSP =====#
ts_ssp = read_ssp('TREFHT', time_span)


#===== read injection amount =====#
PATH = "/glade/campaign/cesm/collections/ARISE-SAI-1.5/b.e21.BW.f09_g17.SSP245-TSMLT-GAUSS-DEFAULT." 
t0, t1, t2, so2 = read_sai(PATH)

# total injection amount
s = so2.sel(year = slice(0, 2070)).sum('loc', keep_attrs=True)

# hemispheric injection diff (NH - SH)
s1 = (so2[:,:,2:].sum('loc', keep_attrs=True) - so2[:,:,:2].sum('loc', keep_attrs=True)).sel(year = slice(0, 2070))

# T0 and T1
t0 = t0.sel(year = slice(0, 2070))
t1 = t1.sel(year = slice(0, 2070))
t2 = t2.sel(year = slice(0, 2070))
```

```python
#============================================#
# Calculate GMST
#============================================#
#===== ARISE =====#
ts_ann = ts.groupby('time.year').mean('time')
# anomaly
ts_a = ts_ann - ts_ann.mean("ens")

# annual mean 
weights = np.cos(np.deg2rad(ts_a.lat))
weights.name = "weights"
gmst = ts_a.weighted(weights).mean(("lat", "lon"))
gmst_indv = ts_ann.weighted(weights).mean(("lat", "lon"))

#plot
# plot_gmst(gmst)

#===== SSP =====#
weights = np.cos(np.deg2rad(ts_ssp.lat))
weights.name = "weights"
gmst_ssp = ts_ssp.groupby('time.year').mean('time').weighted(weights).mean(("lat", "lon"))
```

```python
#============================================#
# how to quantify sai amount? 
#============================================#

#Method1: delta s -> compared to previous year
delta_s = xr.DataArray(np.array(s[:, 1:]) - np.array(s[:, :-1]),
                      dims = ['ens','year'],
                      coords = dict(year=gmst.year,
                                    ens = np.arange(1, 11, 1)))

#Method2: remove ens-mean
s_a = s - s.mean("ens")

# delta s1
delta_s1 = xr.DataArray(np.array(s1[:, 1:]) - np.array(s1[:, :-1]),
                      dims = ['ens','year'],
                      coords = dict(year=gmst.year,
                                    # ens = np.arange(1, 11, 1)))
```

```python
#============================================#
# calculated internal variablity
#============================================#
nino = cal_nino(sst, 1, standardize = True)

tpi = cal_tpi(sst, 1,)

nao, eof = cal_nao(psl, standardize = True)

amo = cal_amo(sst, 1,)

#===== plot IVs =====#

plot_nao(nao, hlt_range=[-0.5, 0.5], title = "std. NAO from SAI")
plot_iv(nino, hlt_range=[-0.7, 0.7], title = "standardized ONI from SAI")
plot_iv(tpi, hlt_range=[-0.7, 0.7], title = "standardized TPI from SAI")
plot_iv(amo, hlt_range=[0, 0], title = "std. AMO from SAI")

```

```python
#============================================#
# save to netcdf files
#============================================#
gmst.assign_coords(ens = np.arange(1,11))
nino.assign_coords(ens=np.arange(1,11))

gmst.to_netcdf('./nc_files/gmst.nc')
nino.to_netcdf('./nc_files/oni.nc')
delta_s.to_netcdf('./nc_files/delta_sai.nc')
nao.to_netcdf('./nc_files/nao.nc')
```

```python
#============================================#
# plot time series of injection amounts
#============================================#

plot_sai(s)
plot_sai(delta_s)
plot_sai(s_a)
plot_gmst(t1)
plot_sai(delta_s1)
```

```python
#============================================#
# Fig1: time series: ONI, GMST, & dS
#============================================#

yr = np.arange(2035, 2070)


import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(20, 12))
outer = gridspec.GridSpec(2, 5, wspace=0.3, hspace=0.25)

for i in range(10):
    inner = gridspec.GridSpecFromSubplotSpec(3, 1,
                    subplot_spec=outer[i], hspace=0.0)
    
    # ONI
    ax1 = plt.Subplot(fig, inner[0])    
    ax1.axhline(y = 0, c = 'k', linestyle='-')
    ax1.axhline(y = 0.5, c = 'k', linestyle='--')
    ax1.axhline(y = -0.5, c = 'k', linestyle='--')
    
    nino_a = nino[i].groupby('time.year').mean('time')
    nino_pos = np.where(nino_a>=0.5, nino_a, 0)
    nino_neg = np.where(nino_a<=-0.5, nino_a, 0)
    nino_wpos = np.where(np.logical_and(nino_a>0, nino_a< 0.5) , nino_a, 0)
    nino_wneg = np.where(np.logical_and(nino_a>-0.5, nino_a< 0.0) , nino_a, 0)
    w = 1.0
    ax1.bar(yr, nino_pos, color="r", alpha=0.4, width = w)
    ax1.bar(yr, nino_neg, color="b", alpha=0.4, width = w)
    # ax1.bar(yr, nino_wpos, color="r", alpha=0.3, width = w)
    # ax1.bar(yr, nino_wneg, color="b", alpha=0.3, width = w)
    ax1.plot(yr, nino_a,c='k')
    
    ax1.set_xlim([2035, 2069])
    ax1.set_xticks([])
    ax1.set_ylim([-1.9, 1.9])
    if i == 0 or i ==5:
        ax1.set_ylabel('ONI', size=15)
        ax1.yaxis.set_label_coords(-0.2, .5)
    ax1.tick_params(axis='y', labelsize=15)
    # else:
    #     ax1.yaxis.set_ticklabels([])

    
    # Anotation
    if i ==9:
        ax1.text(0.0, 1.1, 'Member 010', transform=ax1.transAxes, size=15, weight="bold")
    else:
        ax1.text(0.0, 1.1, 'Member 00'+str(i+1), transform=ax1.transAxes, size=15, weight="bold")

    fig.add_subplot(ax1)

    
    #GMST
    ax2 = plt.Subplot(fig, inner[1])
    #ax2.plot(gmst[i] + 1.5,c='r')
    ax2.axhline(y = 1.5, c = 'k', linestyle='--')
    ax2.plot(yr, gmst_indv[i] - Target, c='blue')
    
    ax2.set_xlim([2035, 2069])
    ax2.set_xticks([])
    
    # ax2.yaxis.tick_right()
    ax2.set_ylim([1.1, 1.9])
    ax2.yaxis.set_ticks(np.arange(1.2, 1.8, 0.2))
    if i == 0 or i ==5:
        ax2.set_ylabel('GMST (K)', size=15, c ='b')
        ax2.yaxis.set_label_coords(-0.2, .5)
        # ax2.yaxis.set_label_position("right")
    ax2.tick_params(axis='y', labelsize=15, colors='b')

    
    fig.add_subplot(ax2)
   

    # delta S
    ax3 = plt.Subplot(fig, inner[2])
    ax3.plot(yr[5:], delta_s[i,5:], c = 'orange')
    ax3.plot(yr[:6], delta_s[i,:6], '--', c = 'orange')
    ax3.axhline(y = 0, c = 'k', linestyle='--')
    
    ax3.set_xlim([2035, 2069])
    ax3.set_ylim([-0.6, 1.19])
    ax3.yaxis.set_ticks(np.arange(-0.5, 1.4, 0.5))
    if i == 0 or i ==5:
        ax3.set_ylabel(u'Δ $SO_{2}$ (Tg)', size=15, c= 'orange')
        ax3.yaxis.set_label_coords(-0.2, .5)
    
    ax3.tick_params(axis='y', labelsize=15, colors = 'orange')

    
    ax3.xaxis.set_ticks(np.arange(2035, 2069, 10))
    ax3.tick_params(axis='x', labelsize=15)
    
    fig.add_subplot(ax3)
    

# plt.savefig('./figs/Fig2.pdf', dpi=300)
```
